# test de normalit√© 

```python
for metric in df_all_segments_stats.columns[2:]:

    print(f" metrique =========== {metric} =============")
    x = df_all_segments_stats[metric]

    # Histogramme noramle si la distribution suis une loi normale 
    plt.hist(x, bins=20, density=True)
    plt.title("Histogramme")
    plt.show()

    # QQ-plot  # normale si les points sont align√©es avec la droite 
    stats.probplot(x, dist="norm", plot=plt)
    plt.title("QQ-plot")
    plt.show()

    # Test de normalit√© : pas normale si p < 0,05
    stat, p = stats.shapiro(x)
    print(f"Shapiro-Wilk p-value {metric} = {p:.4f}")
```
